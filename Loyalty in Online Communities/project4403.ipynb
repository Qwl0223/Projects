Then we can do some calculation. Firstly, we can judge whether a user is loyal to a community. Here is a function to judge whether an user is loyal to a community. The input attribut is username and the name of the community.提示由于数据集只有20个，这边的用户是否忠诚的运行结果可能会与在所有数据集中运行存在差异

```python
def isloyal(username, communityname):
    with open(path + "usertimes.csv",'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row['Username'] == username:
                key_list = []
                value_list = []

                for key, value in row.items():
                    key_list.append(key), value_list.append(value)
                #print(value_list)
                for i in range(0,len(key_list)):
                    key_value = key_list[i]
                    if key_value == communityname:
                        comment_times = value_list[i]
                value_list.pop(0)

                l = list(map(eval, value_list))
                total_comment = np.sum(l)
                
                if total_comment == 0:
                    result = 0
                else:
                    result = float(comment_times)/float(total_comment)
                
                loyal_rate = format(result, '.2f')

    print(username, 'has', comment_times, 'comments in', communityname)
    print(username, 'has', total_comment, 'comments in all the Reddit community')
    print(username,'loyalty to',communityname, 'is',loyal_rate)

    if float(loyal_rate) > 0.5:
        print(username, 'is a loyal user to ', communityname)
    else:
        print(username, 'is not a loyal user to ', communityname)

    return loyal_rate
```

```python
def isloyalcommunity(subreddit):

    userlist = pd.read_csv(path + "usertimes.csv", sep=',')["Username"].values
    counter = 0
    for i in range(0, len(userlist  )):
        username = userlist[i]
        communityname = subreddit
        result = isloyal(username,communityname)
        if result == None:
            results =0
        else:
            results = result    
        result = float(result)
        if result > 0.5:
            counter = counter+1
    print(counter)
    if counter > 25:
        print(subreddit,'is a loyal community')
    else:
        print(subreddit, 'is not a loyal community')
    
    return counter
```

```python
def nodes(subreddit):
    with open(path_dataset + subreddit + ".json", 'r') as load_f:
        load_dict = json.load(load_f)
        nece_dict = load_dict[0]
        # print(nece_dict)
        keys = list(nece_dict.keys())
        #print(len(keys))
        values = list(nece_dict.values())
        #print(len(values))

    with open(path_eachmonprocess + subreddit + '_1.csv', mode='w') as data_file:
        data_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        data_writer.writerow(['username'])
        for i in range (0,len(keys)):
            a = keys[i]
            #print(a)
            data_writer.writerow([a])

    with open(path_eachmonprocess + subreddit + '_2.csv', mode='w') as data_file:
        data_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        data_writer.writerow(['username'])
        for i in range(0, len(values)):
            b = values[i]
            for i in range(0, len(b)):
                c = b[i]
                data_writer.writerow([c])

    with open(path_eachmonprocess + subreddit + '_1.csv', 'ab') as f:
        f.write(open(path_eachmonprocess + subreddit + '_2.csv', 'rb').read())

    with open(path_eachmonuser + subreddit + '.csv', mode='w') as data_file:
        data_writer = csv.writer(data_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        data_writer.writerow(['Username', subreddit])

        username = pd.read_csv(filepath_or_buffer=path_eachmonprocess + subreddit + '_1.csv', sep=',')["username"].values
        bb = set(username)
        for each_bb in bb:
            count = 0
            for each_username in username:
                if each_bb == each_username:
                    count += 1
            # print(each_bb, ": ", count)
            data_writer.writerow([each_bb, count])

    usernamelist = pd.read_csv(filepath_or_buffer=path_eachmonuser + subreddit + '.csv', sep=',')["Username"].values
    #print(usernamelist)
    return usernamelist
```

```python
def edges(subreddit):

    with open( path_dataset + subreddit + ".json", 'r') as load_f:
        load_dict = json.load(load_f)
        nece_dict = load_dict[0]
        #print(nece_dict)
        keys = list(nece_dict.keys())
        #print(len(keys))
        values = list(nece_dict.values())
        #print(len(values))

        for j, user1 in enumerate(keys):
            for s,user2 in enumerate(values[j]):
                yield user1,user2
```

```python
def graph(subreddit):
    G = nx.Graph()
    node = nodes(subreddit)
    G.add_nodes_from(node)
    G.add_edges_from(edges(subreddit))
    return G
```

```python
def degrees(G):
    return [G.degree(u) for u in G]
```
